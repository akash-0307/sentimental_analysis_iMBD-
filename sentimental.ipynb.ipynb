```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url="https://www.imdb.com/title/tt6468322/reviews?ref_=tt_urv"
```

```python
r=requests.get(url)
```

```python
html_content=r.content
```

```python
soup=bs(html_content,"html.parser")
```

```python
soup
```

```python
title=soup.title
```

```python
title.string
```

# Name of reviewer

```python
names=soup.find_all("span",class_="display-name-link")
cust_name=[]
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
```

```python
cust_name
```

# Ratings

```python
star=soup.find_all("span",class_="rating-other-user-rating")

cust_star=[]
for i in range(0,len(star)):
    cust_star.append(star[i].get_text())
```

```python
cust_star
```

```python
anchor_tags = soup.find_all('a', class_='title')

# Extract the text from each anchor tag
cust_text=[]
for i in anchor_tags:
    cust_text.append(i.get_text(strip=True))
    
```

```python
cust_text
```

# Descriptive review

```python
rev=soup.find_all("div",class_="text show-more__control")
cust_rev=[]
for i in range(0,len(rev)):
    cust_rev.append(rev[i].get_text())
    
```

```python
cust_rev
```

# Date

```python
time=soup.find_all("span", class_="review-date")
cust_time=[]
for i in range(0,len(time)):
    cust_time.append(time[i].get_text())
```

```python
cust_time
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
```

```python
df["Date"]=cust_time
df["name"]=cust_name
df["star-rating"]=cust_star
df["review-keyword"]=cust_text
df["descriptive_review"]=cust_rev
```

```python
df
```

```python
df.to_csv("money-heist_iMbD_review.csv")

```

# Sentimental analysis

```python
from textblob import TextBlob
```

```python
polarity=[]
subject=[]

for q in df["descriptive_review"]:
    a=TextBlob(q)
    polarity.append(a.polarity)
    subject.append(a.subjectivity)
```

```python
df["polarity"]=polarity
df["subjectivity"]=subject
```
